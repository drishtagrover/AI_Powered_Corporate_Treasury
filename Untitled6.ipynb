```python
# ==============================
# 1. Imports
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Embedding, LSTM, Dense, Concatenate, Dropout,
    Masking, Flatten, BatchNormalization, Bidirectional
)

# ==============================
# 2. Load Data
# ==============================
file_path = "/content/final.xlsx"
df = pd.read_excel(file_path)

print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())

# ==============================
# 3. Columns Setup
# ==============================
company_col = "Company"
date_col = "FiscalDate"

TARGETS = [
    "Risk_Score",
    "Growth_Score",
    "Profitability_Score",
    "Stability_Score",
    "Sector_Score",
    "Overall_Score",
]

MACRO_FEATURES = ["GST", "CorpTax%", "Inflation%", "RepoRate%", "USDINR_Close"]
SEQ_FEATURES = TARGETS + MACRO_FEATURES

# ==============================
# 4. Handle Missing Values
# ==============================
for col in SEQ_FEATURES:
    df[col] = df.groupby(company_col)[col].transform(lambda g: g.ffill().bfill())
df = df.dropna(subset=SEQ_FEATURES)

# ==============================
# 5. Encode Company
# ==============================
le = LabelEncoder()
df["Company_Encoded"] = le.fit_transform(df[company_col])

# ==============================
# 6. Scaling
# ==============================
scaler_X = StandardScaler()
scaler_y = StandardScaler()
df[SEQ_FEATURES] = scaler_X.fit_transform(df[SEQ_FEATURES])

# ==============================
# 7. Sequence Builder
# ==============================
SEQ_LEN = 16  # try 12 or 20 also

def create_sequences(group):
    X, y, comp = [], [], []
    values = group[SEQ_FEATURES].values
    targets = group[TARGETS].values
    company_id = group["Company_Encoded"].values[0]

    for i in range(len(values) - SEQ_LEN):
        X.append(values[i:i+SEQ_LEN])
        y.append(targets[i+SEQ_LEN])
        comp.append(company_id)

    return np.array(X), np.array(y), np.array(comp)

X_list, y_list, comp_list = [], [], []
for _, g in df.groupby(company_col):
    X_g, y_g, comp_g = create_sequences(g)
    if len(X_g) > 0:
        X_list.append(X_g)
        y_list.append(y_g)
        comp_list.append(comp_g)

X = np.vstack(X_list)
y = np.vstack(y_list)
companies = np.concatenate(comp_list)

print("Sequences shape:", X.shape, y.shape)

# ==============================
# 8. Train/Test Split
# ==============================
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train_raw, y_test_raw = y[:split], y[split:]
comp_train, comp_test = companies[:split], companies[split:]

# Fit y scaler
y_train = scaler_y.fit_transform(y_train_raw)
y_test = scaler_y.transform(y_test_raw)

# ==============================
# 9. BiLSTM Model
# ==============================
num_companies = df["Company_Encoded"].nunique()
embed_dim = 8

input_seq = Input(shape=(SEQ_LEN, len(SEQ_FEATURES)))
input_comp = Input(shape=(1,))

# Company embedding
x_comp = Embedding(input_dim=num_companies, output_dim=embed_dim)(input_comp)
x_comp = Flatten()(x_comp)

# Sequence path
x_seq = Masking()(input_seq)
x_seq = Bidirectional(LSTM(128, return_sequences=True))(x_seq)
x_seq = Dropout(0.2)(x_seq)
x_seq = Bidirectional(LSTM(64, return_sequences=True))(x_seq)
x_seq = Dropout(0.2)(x_seq)
x_seq = Bidirectional(LSTM(32))(x_seq)
x_seq = Dropout(0.2)(x_seq)

x = Concatenate()([x_seq, x_comp])

dense1 = Dense(64, activation="relu")(x)
dense1 = BatchNormalization()(dense1)
dense1 = Dropout(0.2)(dense1)

dense2 = Dense(32, activation="relu")(dense1)
dense2 = BatchNormalization()(dense2)
dense2 = Dropout(0.2)(dense2)

output = Dense(len(TARGETS))(dense2)

model = Model(inputs=[input_seq, input_comp], outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005), loss="mae")

model.summary()

# ==============================
# 10. Training
# ==============================
es = tf.keras.callbacks.EarlyStopping(patience=12, restore_best_weights=True)
rlr = tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=6, min_lr=1e-5)

history = model.fit(
    [X_train, comp_train],
    y_train,
    validation_split=0.2,
    epochs=120,
    batch_size=32,
    verbose=1,
    callbacks=[es, rlr]
)

# ==============================
# 11. BiLSTM Evaluation
# ==============================
preds_scaled = model.predict([X_test, comp_test])
preds = scaler_y.inverse_transform(preds_scaled)

print("\nBiLSTM MAE per Score:")
for i, col in enumerate(TARGETS):
    mae = mean_absolute_error(y_test_raw[:, i], preds[:, i])
    print(f"{col}: {mae:.4f}")

# ==============================
# 12. RF and XGB Ensemble
# ==============================
# Get BiLSTM predictions for train and test
train_preds_scaled = model.predict([X_train, comp_train])
train_preds = scaler_y.inverse_transform(train_preds_scaled)

test_preds_scaled = model.predict([X_test, comp_test])
test_preds = scaler_y.inverse_transform(test_preds_scaled)

X_train_meta, y_train_meta = train_preds, y_train_raw
X_test_meta, y_test_meta = test_preds, y_test_raw

# ---- Random Forest Ensemble ----
rf_preds = []
for i in range(len(TARGETS)):
    rf = RandomForestRegressor(n_estimators=300, random_state=42)
    rf.fit(X_train_meta, y_train_meta[:, i])
    rf_pred = rf.predict(X_test_meta).reshape(-1, 1)
    rf_preds.append(rf_pred)

rf_preds = np.hstack(rf_preds)

print("\nRandom Forest Ensemble MAE per Score:")
for i, col in enumerate(TARGETS):
    mae = mean_absolute_error(y_test_meta[:, i], rf_preds[:, i])
    print(f"{col}: {mae:.4f}")

# ---- XGBoost Ensemble ----
xgb_preds = []
for i in range(len(TARGETS)):
    xgb = XGBRegressor(
        n_estimators=400, learning_rate=0.05,
        max_depth=6, subsample=0.8, colsample_bytree=0.8,
        random_state=42
    )
    xgb.fit(X_train_meta, y_train_meta[:, i])
    xgb_pred = xgb.predict(X_test_meta).reshape(-1, 1)
    xgb_preds.append(xgb_pred)

xgb_preds = np.hstack(xgb_preds)

print("\nXGBoost Ensemble MAE per Score:")
for i, col in enumerate(TARGETS):
    mae = mean_absolute_error(y_test_meta[:, i], xgb_preds[:, i])
    print(f"{col}: {mae:.4f}")

# ==============================
# 13. Visualization (BiLSTM only)
# ==============================
n_plot = 200
plt.figure(figsize=(14, 8))
for i, col in enumerate(TARGETS):
    plt.subplot(2, 3, i+1)
    plt.plot(y_test_raw[:n_plot, i], label="Actual")
    plt.plot(preds[:n_plot, i], label="BiLSTM Predicted")
    plt.title(col)
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ==============================
# 1. Imports
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Embedding, LSTM, Dense, Concatenate, Dropout,
    Masking, Flatten, BatchNormalization, Bidirectional
)

# ==============================
# 2. Load Data
# ==============================
file_path = "/content/final.xlsx"
df = pd.read_excel(file_path)

print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())

# ==============================
# 3. Columns Setup
# ==============================
company_col = "Company"
date_col = "FiscalDate"

TARGETS = [
    "Risk_Score",
    "Growth_Score",
    "Profitability_Score",
    "Stability_Score",
    "Sector_Score",
    "Overall_Score",
]

MACRO_FEATURES = ["GST", "CorpTax%", "Inflation%", "RepoRate%", "USDINR_Close"]
SEQ_FEATURES = TARGETS + MACRO_FEATURES

# ==============================
# 4. Handle Missing Values
# ==============================
for col in SEQ_FEATURES:
    df[col] = df.groupby(company_col)[col].transform(lambda g: g.ffill().bfill())
df = df.dropna(subset=SEQ_FEATURES)
# ==============================
# 4.5 Outlier Capping (Winsorization)
# ==============================
def cap_outliers(series, lower_quantile=0.01, upper_quantile=0.99):
    """
    Cap outliers at the given quantiles (default: 1st and 99th percentile).
    """
    lower = series.quantile(lower_quantile)
    upper = series.quantile(upper_quantile)
    return np.clip(series, lower, upper)

for col in SEQ_FEATURES:
    df[col] = cap_outliers(df[col])


# ==============================
# 5. Encode Company
# ==============================
le = LabelEncoder()
df["Company_Encoded"] = le.fit_transform(df[company_col])

# ==============================
# 6. Scaling
# ==============================
scaler_X = StandardScaler()
scaler_y = StandardScaler()
df[SEQ_FEATURES] = scaler_X.fit_transform(df[SEQ_FEATURES])

# ==============================
# 7. Sequence Builder
# ==============================
SEQ_LEN = 16  # you can try 12 or 20 as well

def create_sequences(group):
    X, y, comp = [], [], []
    values = group[SEQ_FEATURES].values
    targets = group[TARGETS].values
    company_id = group["Company_Encoded"].values[0]

    for i in range(len(values) - SEQ_LEN):
        X.append(values[i:i+SEQ_LEN])
        y.append(targets[i+SEQ_LEN])
        comp.append(company_id)

    return np.array(X), np.array(y), np.array(comp)

X_list, y_list, comp_list = [], [], []
for _, g in df.groupby(company_col):
    X_g, y_g, comp_g = create_sequences(g)
    if len(X_g) > 0:
        X_list.append(X_g)
        y_list.append(y_g)
        comp_list.append(comp_g)

X = np.vstack(X_list)
y = np.vstack(y_list)
companies = np.concatenate(comp_list)

print("Sequences shape:", X.shape, y.shape)

# ==============================
# 8. Train/Test Split
# ==============================
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train_raw, y_test_raw = y[:split], y[split:]
comp_train, comp_test = companies[:split], companies[split:]

# Fit y scaler on training targets and transform
y_train = scaler_y.fit_transform(y_train_raw)
y_test = scaler_y.transform(y_test_raw)

# ==============================
# 9. BiLSTM Model
# ==============================
num_companies = df["Company_Encoded"].nunique()
embed_dim = 8

input_seq = Input(shape=(SEQ_LEN, len(SEQ_FEATURES)))
input_comp = Input(shape=(1,))

# Company embedding
x_comp = Embedding(input_dim=num_companies, output_dim=embed_dim)(input_comp)
x_comp = Flatten()(x_comp)

# Sequence path
x_seq = Masking()(input_seq)
x_seq = Bidirectional(LSTM(128, return_sequences=True))(x_seq)
x_seq = Dropout(0.2)(x_seq)
x_seq = Bidirectional(LSTM(64, return_sequences=True))(x_seq)
x_seq = Dropout(0.2)(x_seq)
x_seq = Bidirectional(LSTM(32))(x_seq)
x_seq = Dropout(0.2)(x_seq)

x = Concatenate()([x_seq, x_comp])

dense1 = Dense(64, activation="relu")(x)
dense1 = BatchNormalization()(dense1)
dense1 = Dropout(0.2)(dense1)

dense2 = Dense(32, activation="relu")(dense1)
dense2 = BatchNormalization()(dense2)
dense2 = Dropout(0.2)(dense2)

output = Dense(len(TARGETS))(dense2)

model = Model(inputs=[input_seq, input_comp], outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005), loss="mae")

model.summary()

# ==============================
# 10. Training
# ==============================
es = tf.keras.callbacks.EarlyStopping(patience=12, restore_best_weights=True)
rlr = tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=6, min_lr=1e-5)

history = model.fit(
    [X_train, comp_train],
    y_train,
    validation_split=0.2,
    epochs=120,
    batch_size=32,
    verbose=1,
    callbacks=[es, rlr]
)

# ==============================
# 11. Evaluation
# ==============================
preds_scaled = model.predict([X_test, comp_test], verbose=0)
preds = scaler_y.inverse_transform(preds_scaled)

# --- Helper: scale values to 0–100
def scale_to_100(values, col, df):
    col_min, col_max = df[col].min(), df[col].max()
    return 100 * (values - col_min) / (col_max - col_min)

print("\nEvaluation Metrics (0–100 scale):")
for i, col in enumerate(TARGETS):
    y_true_100 = scale_to_100(y_test_raw[:, i], col, df)
    y_pred_100 = scale_to_100(preds[:, i], col, df)
    mae = mean_absolute_error(y_true_100, y_pred_100)
    print(f"{col}: MAE={mae:.2f}")

# ==============================
# 12. Visualization
# ==============================
n_plot = 200
plt.figure(figsize=(14, 8))
for i, col in enumerate(TARGETS):
    plt.subplot(2, 3, i+1)
    y_true_100 = scale_to_100(y_test_raw[:n_plot, i], col, df)
    y_pred_100 = scale_to_100(preds[:n_plot, i], col, df)
    plt.plot(y_true_100, label="Actual (0–100)")
    plt.plot(y_pred_100, label="Predicted (0–100)")
    plt.title(col)
    plt.legend()
plt.tight_layout()
plt.show()

# ==============================
# 13. Predict Future for a Company
# ==============================
def predict_future(company_name, df, model, scaler_X, scaler_y, SEQ_LEN=16):
    if company_name not in df[company_col].unique():
        raise ValueError(f"Company '{company_name}' not found in dataset.")

    company_data = df[df[company_col] == company_name].sort_values(by=date_col)
    company_id = int(company_data["Company_Encoded"].iloc[0])

    if len(company_data) < SEQ_LEN:
        raise ValueError(f"Not enough data for {company_name}. Need at least {SEQ_LEN} quarters.")

    last_seq = company_data[SEQ_FEATURES].values[-SEQ_LEN:]
    last_seq = np.expand_dims(last_seq, axis=0)

    company_input = np.array([[company_id]], dtype=np.int32)

    pred_scaled = model.predict([last_seq, company_input], verbose=0)
    pred = scaler_y.inverse_transform(pred_scaled)[0]

    # Rescale each prediction to 0–100 based on dataset min/max
    scaled_pred = {}
    for i, col in enumerate(TARGETS):
        col_min, col_max = df[col].min(), df[col].max()
        scaled_val = 100 * (pred[i] - col_min) / (col_max - col_min)
        scaled_pred[col] = round(float(scaled_val), 2)

    return scaled_pred

# ==============================
# 14. Example Predictions
# ==============================
example_companies = df[company_col].unique()[:3]
for comp in example_companies:
    print(f"\nNext quarter prediction for {comp} (0–100 scale):")
    print(predict_future(comp, df, model, scaler_X, scaler_y))
```
